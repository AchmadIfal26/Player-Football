# Added category age based on experience

```python
# Determine the age category

def categorize_age(age):
    if age < 25:
        return "Young"
    elif 25 <= age < 30:
        return "Prime"
    elif 30 <= age < 35:
        return "Experienced"
    else:
        return "Veteran"
```

```python
# Added age category column
player['age_category'] = player['age'].apply(categorize_age)
```

# Data Analysis

# 1. Player Position Distribution

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Use a Seaborn style for the plot
sns.set(style="whitegrid")

# Set the figure size (width, height) in inches
plt.figure(figsize=(12, 8))

# Create the plot
position_counts = player['position'].value_counts(ascending=True)
ax = position_counts.plot(kind='bar', color='skyblue', edgecolor='black')

# Add titles and labels
plt.title('Distribution of Total Players', fontsize=14)
plt.xlabel('Position', fontsize=12)
plt.ylabel('Total Players', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Add grid lines
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add value labels on the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

# Show the plot
plt.tight_layout()
plt.show()
```

# 2. Average Player Height by Position

```python
average_height = player.groupby('position')['height'].mean()
print(average_height)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the average height per position
average_height = player.groupby('position')['height'].mean()

# Sort the average_height Series by values
average_height_sorted = average_height.sort_values()

# Use a Seaborn style for the plot
sns.set(style="whitegrid")

# Set the figure size (height, width) in inches for horizontal plot
plt.figure(figsize=(10, 8))

# Create the plot
ax = average_height_sorted.plot(kind='barh', color='lightgreen', edgecolor='black')

# Add titles and labels
plt.title('Average Height by Position', fontsize=14)
plt.xlabel('Average Height (cm)', fontsize=12)
plt.ylabel('Position', fontsize=12)

# Add grid lines
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Add value labels on the bars
for p in ax.patches:
    ax.annotate(f'{p.get_width():.2f}', (p.get_width(), p.get_y() + p.get_height() / 2.),
                ha='left', va='center', xytext=(10, 0), textcoords='offset points')

# Show the plot
plt.tight_layout()
plt.show()
```

# 5. Distribution of the Player's Dominant Foot

```python
# Calculate the distribution of player's dominant foot
foot_distribution = player['foot'].value_counts()

# Define custom colors for the pie chart
colors = ['#ff9999', '#66b3ff', '#99ff99']

# Define explode parameter to highlight a slice
explode = (0.1, 0, 0)  # explode the first slice (dominant foot)

# Create the pie chart
plt.figure(figsize=(8, 8))
foot_distribution.plot(kind='pie', autopct='%1.1f%%', colors=colors, explode=explode, shadow=True)

# Add title
plt.title("Distribution of Player's Dominant Foot")

# Add legend
plt.legend(foot_distribution.index, loc='upper right')

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')

# Show the plot
plt.show()
```

# 7. Analyze Clubs with the Highest Players

```python
average_height_by_club = player.groupby('club')['height'].mean()
tallest_club = average_height_by_club.max()
print(f"Klub dengan rata-rata tinggi pemain tertinggi adalah: {tallest_club}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the average height per club
average_height_by_club = player.groupby('club')['height'].mean()

# Determine the club with the tallest players
tallest_club = average_height_by_club.idxmax()

print(f"Club with the tallest average players' height: {tallest_club}")

# Use a Seaborn style for the plot
sns.set(style="whitegrid")

# Set the figure size (width, height) in inches
plt.figure(figsize=(12, 8))

# Create the bar plot
ax = average_height_by_club.plot(kind='bar', color='skyblue', edgecolor='black')

# Highlight the club with the tallest players
ax.bar(tallest_club, average_height_by_club[tallest_club], color='red')

# Add title and labels
plt.title('Average Height of Players by Club', fontsize=14)
plt.xlabel('Club', fontsize=12)
plt.ylabel('Average Height (cm)', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Add grid lines
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add numeric values on top of each bar (rotated and enlarged)
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height:.2f}', (p.get_x() + p.get_width() / 2., height),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points', rotation=45, fontsize=10)

# Show the plot
plt.tight_layout()
plt.show()
```
